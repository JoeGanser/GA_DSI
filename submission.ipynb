### Using t-tests & bootstrapping for feature selection

In these t-tests, we will test to see if the mean of each feature is different in the malignant cases and the cancerous cases. If the average is different, we assume the feature is useful in predicting cancer.

** False positives/negatives in the t-tests **

Sometimes hypothesis tests are wrong. To accomodate for this, we'll re-run of the hypothesis test on 1000 random small subsets of each feature, and see what percentage of the time we can reject the null hypothesis. If we reject the null hypothesis in more than 900 t-tests, we'll be pretty confident.

##### NULL HYPOTHESIS: Means of the feature are the same in both CANCER and BENIGN cases (useless feature)
##### ALTERNATIVE: Means of the feature are different in cancer and non-cancer cases. (useful feature)

```python
import numpy as np
from scipy.stats import t
from scipy import stats
import random

def t_test(x,y,P=0.99):
    #assume unequal variance, degrees of freedom
    t_test,p_value = stats.ttest_ind(x,y,equal_var = False)
    #calculate degrees of freedom using Welch's formula
    dof = (np.var(x)/len(x) + np.var(y)/len(y))**2 / ((np.var(x)/len(x))**2 / (len(x)-1) + (np.var(y)/len(y))**2 / (len(y)-1))
    dof = round(dof)
    #specify crticial bounds
    t_critical_lower = t.ppf(P, dof)
    t_critical_upper = t.ppf(1-P,dof)
    if (t_test>t_critical_upper)&(p_value<1-P):
        return "Reject Null"
    elif (t_test<t_critical_lower)&(p_value<1-P):
        return "Reject Null"
    else:
        return "Fail to reject"
    
features = [i for i in df.columns if i not in ['ID','diagnosis']]
N_tests = 1000
test_results = {}
for f in features:
    _ = []
    M = list(df[df['diagnosis']=='M'][f])
    B = list(df[df['diagnosis']=='B'][f])
    for i in range(N_tests):
        Mx = random.sample(M,29)
        By = random.sample(B,29)
        _.append(t_test(Mx,By))
    test_results[f] = {i:100*_.count(i)/len(_) for i in set(_)}
results = pd.DataFrame(test_results).transpose()
results.columns = ['Percent_Fail_to_reject','Percent_Reject_Null']
def conclusion(alpha,beta):
    if beta>70:
        return 'useful in prediction'
    elif alpha>70:
        return 'NOT useful'
    else:
        return 'possibly useful'
results['useful_in_predicting'] = results.apply(lambda x: conclusion(x.Percent_Fail_to_reject,x.Percent_Reject_Null),axis=1)
results
```

The frame above shows the break down of what percentage of the time we rejected the null hypothesis for t-tests on 1000 random samples of each feature from the malginant vs benign classes. 

Hence for the feature of radius mean, 100% of all t-tests on 29 random samples of benign mean radius versus cancerous mean radius lead us to the conclusion that the average mean radius is DIFFERENT for cancerous and benign cases.

If I were to pick 3 useful predictors, they would be:
* radius_mean
* radius_worst
* texture_mean

Going forward, it would be best to include only features that are both useful and possibly useful in modelling.

# Modelling: Logistic Regression versus Random Forest

Machine learning is a highly iterative science. Thus, when solving a classification process, its always best to start with a simple model to see how it performs. Starting with logistic regression, we can get a baseline of performance and then compare it to random forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score,f1_score,roc_curve,auc
from sklearn.linear_model import LogisticRegression

X = df.drop('diagnosis',axis=1)
y = df['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


lr = LogisticRegression()
lr.fit(X_train,y_train)
y_prob = lr.predict_proba(X_test)[:,1]
y_pred = lr.predict(X_test)
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(false_positive_rate, true_positive_rate)
plt.title('Log Reg: Receiver Operating Characteristic')
plt.plot(false_positive_rate,true_positive_rate, color='red',label = 'AUC = %0.2f' % roc_auc)
plt.plot([0, 1], [0, 1],linestyle='--')
plt.axis('tight')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
print(classification_report(y_test, y_pred))
print('Accuracy on test set: ',accuracy_score(y_test, y_pred))
print('\n')
print("confusion matrix for test set:")
print(confusion_matrix(y_test, y_pred))
print('Confusion matrix for train set: ',)
print(confusion_matrix(y_train, lr.predict(X_train)))
print('Accuracy for train set: ',accuracy_score(y_train, lr.predict(X_train)))
plt.show()
```

## Comments on logistic regression:
    
This is generally pretty good performance despite there being no hyper parameter tuning. Although there is slight discrepancy between the ratios of false positives and false negatives between the train and test sets, the model performs well. One possible way to correct this could be through using class balancing techniques in the original data set, such as SMOTE.

We can examine the magnitude of influence on the coefficients. The absolute value of the product of each feature's standard deviation multiplied by the logisitic regression coefficient can give us ranking of feature influence.

```python
coeffs = pd.DataFrame({i:[j] for i,j in zip(list(X.columns),list(lr.coef_[0]))}).transpose()
coeffs = pd.concat([coeffs,df.drop('diagnosis',axis=1).std(axis=0)],axis=1)
coeffs.columns = ['log_reg_coef','feature_standard_dev']
coeffs['influence'] = coeffs['log_reg_coef']*coeffs['feature_standard_dev']
coeffs
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier()
rf_params = {'n_estimators':[40,45,50,60,65,75],'criterion':['gini','entropy'],'class_weight':['balanced',None]}

GS = GridSearchCV(rf,rf_params)
GS.fit(X,y)
print("best random forest parameters: ",GS.best_params_)
print("best random forest score: ",GS.best_score_)
```

Random forest performs very well, even better than logistic regression. As we can see the false positive and false negative rates are about equal to each other on both the train set and the test set. This does indicate a good fit. Issues regarding over/under fitting are accounted for by hyperparameter tuning, inconjunction to the fact that random forest is generally less likely to over fit.
